# Part 1: Random Image Generation

```
import IPython.display
import PIL.Image
import numpy as np
import cv2
 
# A simple function to display an image in an ipython notebook
def nbimage( data ):
  IPython.display.display(PIL.Image.fromarray(data))
 
 
def draw_random_circle(img):
  center = np.random.uniform(low=np.array([0]*2), high=np.array([512]*2)).astype(np.uint16)
  radius = np.random.uniform(low=np.array([0]), high=np.array([50])).astype(np.uint16)
  color = np.random.uniform(low=np.array([0]*3), high=np.array([255]*3)).astype(np.uint8)
  cv2.circle(img, tuple(center), radius=radius, color=tuple(color.tolist()), thickness=-1)
#   cv2.circle(img, (0,0), 10, (128,128,128), -1)

# create an image consisting of random colors
N_shapes = 300
img = np.zeros((512,512,3), dtype=np.uint8) # a 512x512 image, with 3 color channels (R,G,B)
 
for i in range(N_shapes):
  draw_random_circle(img)
 
# display it!
nbimage( img )
```
